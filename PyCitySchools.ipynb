```python
# count the number of schools and students
total_number_of_schools = len(student_school_df["School ID"].unique())
total_number_of_students = len(student_school_df["Student ID"].unique())

print ('Total Schools: '+ str(total_number_of_schools))
print ('Total Students: '+ str(total_number_of_students))
```

```python
# Find the total budget
total_budget = school_df["budget"].sum()
print ("Total Budget: " + str(total_budget))
```

```python
# find the average math score
average_math_score = round(student_df["math_score"].mean(),2)
average_math_score
average_reading_score = round(student_df["reading_score"].mean(),2)
print ('average math score: ' + str(average_math_score) +' average reading score: ' + str(average_reading_score))
```

```python
# Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
overall_passing_rate = round((average_math_score + average_reading_score)/2,2)
overall_passing_rate
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
# create dataframe if math score >=70 then true, else false
passing_math = student_df["math_score"] >= 70
passing_math
# find the mean() 
percent_passing_math = round((passing_math.mean())*100,2)
percent_passing_math
```

```python
# find percent passing reading
passing_reading = student_df["reading_score"] >=70
passing_reading
# find the mean()
percent_passing_reading = round((passing_reading.mean())*100,2)
percent_passing_reading
```

# School Summary

#Create an overview table that summarizes key metrics about each school, including:

* School Name
* School Type
* Total Students
* Total School Budget
* Per Student Budget
* Average Math Score
* Average Reading Score
* % Passing Math (The percentage of students that passed math.)
* % Passing Reading (The percentage of students that passed reading.)
* % Overall Passing (The percentage of students that passed math and reading.)

```python
# start with all students and schools
student_school_df
# add passing math and reading
student_school_df["passing_math"] = student_school_df["math_score"] >= 70
student_school_df["passing_reading"] = student_school_df["reading_score"] >=70

# add overall passing - if student passed reading and writing then yes else no
def f(row):
    if row['passing_math'] == True & row['passing_reading'] == True:
        val = True
    else:
        val = False
    return val
student_school_df["overall_passing"] = student_school_df.apply(f,axis=1)
# make bool
student_school_df["overall_passing"]=student_school_df["overall_passing"].astype('bool')

student_school_df
math_scores_by_grade = student_school_df.copy()
```

```python
# Reorganizing the columns using double brackets

organized_df = student_school_df[['school_name','type', 'size','budget','math_score','reading_score','passing_math','passing_reading','overall_passing']]
# add per student budget column
organized_df.insert(4,'budget_per_student',organized_df["budget"]/organized_df["size"],True)
# # Rename columns for future report
organized_df.columns=['School Name','School Type','Total Students','Total School Budget','Per Student Budget','Average Math Score','Average Reading Score','% Passing Math','% Passing Reading','% Overall Passing']

student_school_df = organized_df
student_school_df
```

```python
#create new df to hold group of schools
school_group_df = student_school_df.groupby(["School Name"]).mean()
 
# find passing scores a
school_group_df["% Passing Math"] = round(school_group_df["% Passing Math"]*100,2)
school_group_df["% Passing Reading"] = round(school_group_df["% Passing Reading"]*100,2)
school_group_df["% Overall Passing"] = round(school_group_df["% Overall Passing"] *100,2)

# Format data
school_group_df["Total Students"] = school_group_df["Total Students"].map("{:,.0f}".format)
school_group_df["Total School Budget"] = school_group_df["Total School Budget"].map("${:,.2f}".format)
school_group_df["Per Student Budget"] = school_group_df["Per Student Budget"].map("${:,.2f}".format)

school_group_df["Average Math Score"] = school_group_df["Average Math Score"].map("{:,.2f}".format)
school_group_df["Average Reading Score"] = school_group_df["Average Reading Score"].map("{:,.2f}".format)

#passed_math_and_reading
school_group_df
```
